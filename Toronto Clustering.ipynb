```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## I am not familiar with Toronto so my analysis is going to be to visualize where these postal codes are. I am going to cluster these postal codes based on their phsical proximity to each other. 

## Looking at the postal codes, I do not see a clear pattern between the numbers and the letters. Using the latitude and longitude coordinates for the above part two, I will see which postal codes are nearest each other.

```python
X=df_final
```

```python
K_clusters = range(1,10)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]
Y_axis = df_final[['Latitude']]
X_axis = df_final[['Longitude']]
score = [kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(kmeans))]
```

# First I will find the appropriate number of clusters. I will use an 'elbow curve' to see where the addition of another cluster adds less explanatory power.

```python
plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

#  Three clusters looks like a good number without overfitting. 

```python
kclusters=3
kmeans = KMeans(n_clusters =kclusters, init ='k-means++')
kmeans.fit(X[X.columns[1:3]])
X['cluster_label'] = kmeans.fit_predict(X[X.columns[2:3]])
centers = kmeans.cluster_centers_ # Coordinates of cluster centers.
labels = kmeans.predict(X[X.columns[2:3]]) # Labels of each point
X.head(10)
```

```python
X.reset_index(inplace = True)
df_final_clust = df_final.merge(X, left_on='Postcode', right_on='Postcode')
```

```python
X=X[['Postcode', 'cluster_label']]
df_final_clust = df_final.merge(X, left_on='Postcode', right_on='Postcode')
df_final_clust.head()
```

```python
del df_final_clust['cluster_label_y']
```

```python
df_final_clust.rename(columns={'cluster_label_x':'ClusterLabel'}, inplace =True)
```

```python
df_final_clust.head()
```

# Now let's visualize our postal codes.

```python
from folium.plugins import MarkerCluster
```

```python
tor_map = folium.Map(location=[43.4532, -79.3832], zoom_start=10)


x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]


for lat, lng, borough, neighborhood, postcode, ClusterLabel in zip(df_final_clust['Latitude'], df_final_clust['Longitude'],df_final_clust['Borough'], df_final_clust['Neighborhood'], df_final_clust[
'Postcode'], df_final_clust['ClusterLabel']):

    label = '{}, {}, {}'.format(ClusterLabel, postcode, neighborhood)

    label = folium.Popup(label, parse_html=True)

    folium.CircleMarker(

[lat, lng],

radius=5,

popup=label,

color=rainbow[ClusterLabel-1],

fill=True,

fill_color='#3186cc',

fill_opacity=0.7).add_to(tor_map)

tor_map
```
